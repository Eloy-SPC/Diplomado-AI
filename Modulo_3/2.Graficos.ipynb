```python
experiencia = [1,3,5,7,9,13]
salary_data_scientis=[512,624,734,888,1000,9999]
salary_ing_soft=[5212,6224,7234,8828,10200,9929]
```

```python
plt.plot(experiencia,salary_data_scientis)
plt.plot(experiencia,salary_ing_soft)
plt.show
```

```python
plt.plot(experiencia,salary_data_scientis)
plt.plot(experiencia,salary_ing_soft)
plt.title('Salario')
plt.xlabel('experiencia')
plt.ylabel('ganancia')
plt.legend(['Data Scientis', 'Ing Software'])
plt.show()
```

```python
plt.plot(experiencia,salary_data_scientis,color='r', label = 'Data Scientis',linestyle = '--', marker='o')
plt.plot(experiencia,salary_ing_soft,color='g', label = 'Ing Software', linestyle = ':',marker='*')
plt.title('Salario')
plt.xlabel('experiencia')
plt.ylabel('ganancia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Grafico de dispersion

import numpy as np
```

```python
x = np.random.rand(100)
y=np.random.rand(100)
```

```python
plt.scatter(x,y, color='r', alpha=0.1)
plt.title('Gradico de dispersion')
plt.show()
```

```python
#HISTOGRAMA
datos= np.random.randn(10000)#Conjunto de datos aleatorio con una distribucion normal
plt.hist(datos,bins=10)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('histograma')
plt.show()
```

```python
#Grafico de caja y bigotes

datos=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(datos,vert=True)
plt.show
```

```python
#Series de tiempo, grafica conforme alas fechas
import pandas as pd 
import numpy as np
```

```python
data_range= pd.date_range(start='2023-01-01', end='2023-12-01', freq='D')
precios=np.cumsum(np.random.rand(len(data_range))) #Suma acumulativa
```

```python
#Generar series de tiempo
data = pd.DataFrame({'Date':data_range,'precios': precios})
```

```python
data
```

```python
plt.plot(data['Date'], data['precios'], label='precios')
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
categorias=['a','b','c','d']
valores = [5,7,3,8]
datos_bigote=[np.random.normal(0,std,100) for std in range(1,4)]
x3 = np.random.rand(100)
y3=np.random.rand(100)
```

```python
#Cuadricula de figuras
fig, axs= plt.subplots(2,2, figsize = (15,10))
axs[0,0].plot(x,y1)
axs[0,0].set_title('Grafico de lineas')


axs[0,1].bar(categorias,valores)
axs[0,1].set_title('Grafico de barras')

axs[1,1].boxplot(datos_bigote,vert=True)
axs[1,1].set_title('Grafico de cajas')

axs[1,0].scatter(x3,y3, color='r', alpha=0.1)
axs[1,0].set_title('Grafico de dispersion')

plt.show()
```
