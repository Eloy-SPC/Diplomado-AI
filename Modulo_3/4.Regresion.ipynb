# Modelo de regresion simple



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
salarios = pd.read_csv('Salary.csv')
salarios.head()
```

```python
salarios.describe()
```

```python
salarios.info()
```

**Pregunra a resolver**: Determinar cual es el salario de un trabajador basado en el numero de anos de experiencia

```python
#detectar outlier

salarios.skew()
```

```python
import seaborn as sns
```

```python
#Graficamos
sns.histplot(x=salarios['Salary'])
```

```python
sns.histplot(x=salarios['YearsExperience'])
```

```python
sns.scatterplot(x=salarios['YearsExperience'],y=salarios['Salary'])
```

```python
model=LinearRegression()
```

```python
X = salarios[['YearsExperience']]
y= salarios['Salary']
```

```python
#Entrenar modelo
lineal_reg=model.fit(X,y)
```

```python
lineal_reg.coef_
```

```python
lineal_reg.intercept_
```

### El modelo seria 28859.52 + 8731.9410 *Años de exp

```python
#Prediccion para una persona en base a sus anos de experiecia

y_pred= lineal_reg.predict([[2]])
y_pred
```

```python
#Asignar dos para graficar la linea de tendencia 
y_pred=lineal_reg.predict(X)
```

```python
y_pred_m=lineal_reg.intercept_+lineal_reg.coef_*5
y_pred_m
```

```python
#R2 aproximacion de datos
lineal_reg.score(X,y)
```

```python
#graficar

plt.scatter(X,y)
plt.plot(X,y_pred)
```
