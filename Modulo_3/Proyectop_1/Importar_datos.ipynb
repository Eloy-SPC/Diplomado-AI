### Tratar con datos nominales y ordinales

```python
for name in df.columns.to_list():
    print(f'{name} : {df[name].unique()}')
    print('\n')
```

```python
#Forma 1 de remplazar valores
def attrition(s):
    if s == 'Yes':
        return 1
    else:
        return 0
df_clean['Attrition']=df_clean['Attrition'].apply(attrition)
#Forma 2 de remplazar valores
#df_clean['Attrition']=df_clean['Attrition'].apply(lambda s: 1 if s== 'Yes' else 0)
```

```python
#Variables categoricas ordinales 

Attrition_label={'Yes':1, 'No':0}
#pasa la variable con la que quiere cambiar valor
df_clean['Attrition']=df_clean['Attrition'].map(Attrition_label)

```

# Deteccion de valores faltantes


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  
```

```python
#Calcular valores maximos y minimos *quantiles VALORES ATIPICOS
Q1= df_clean['Age'].quantile(0.25)
Q3= df_clean['Age'].quantile(0.75)

IQR = Q3-Q1

limite_inferior = Q1 - 1.5 * IQR

limite_inferior

limite_superior= Q3 +1.5* IQR

limite_superior
```

## Notas: Simulacion de datos atipicos 

```python
#Simular datos atipicos
df_test=df_clean.copy()

indice = np.random.choice(df_test.index,2, replace=False)

df_test.loc[indice[0],'Age'] += 3 * IQR
df_test.loc[indice[1],'Age'] += 3 * IQR
```

```python
df_test.boxplot(column='Age')
```

```python
#Calcular valores maximos y minimos *quantiles VALORES ATIPICOS
Q1= df_test['Age'].quantile(0.25)
Q3= df_test['Age'].quantile(0.75)

IQR = Q3-Q1

limite_inferior = Q1 - 1.5 * IQR

print(limite_inferior)

limite_superior= Q3 +1.5* IQR

print(limite_superior)
```
