## 2 Grafico de variables


```python
def plot_time_series(df, variables, ylabel, title, start_date=None, end_date=None):
    plt.figure(figsize=(15, 5))
    
    if start_date and end_date:
        df = df.loc[start_date:end_date]
    
    for variable in variables:
        if variable in df.columns:
            plt.plot(df.index, df[variable], label=variable)
        else:
            print(f"Columna '{variable}' no encontrada en el DataFrame")
    
    plt.xlabel('Fecha')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_time_series(df_contaminacion,['CO'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['NO'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['NO2'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['O3'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['PM10'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['PM25'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
plot_time_series(df_contaminacion,['SO2'],'COTAMINACION','Agentes contaminantes',start_date='2016-01-01', end_date='2018-12-30')
```

```python
Q1= df_contaminacion['CO'].quantile(0.25)
Q3= df_contaminacion['CO'].quantile(0.75)

IQR = Q3-Q1

limite_inferior = Q1 - 1.5 * IQR
print(limite_inferior)
limite_superior= Q3 +1.5* IQR
print(limite_superior)

df_contaminacion['CO'] = df_contaminacion['CO'].astype(float)

outliers = (df_contaminacion['CO'] < (Q1 - 1.5 * IQR)) | (df_contaminacion['CO'] > (Q3 + 1.5 * IQR))
df_contaminacion.loc[outliers, 'CO'] = np.nan
```
